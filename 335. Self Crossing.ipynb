# 684. Redundant Connection

You are given an array x of n positive numbers. You start at point (0,0) and moves x[0] metres to the north, then x[1] metres to the west, x[2] metres to the south, x[3] metres to the east and so on. In other words, after each move your direction changes counter-clockwise.

Write a one-pass algorithm with O(1) extra space to determine, if your path crosses itself, or not.

## Terrible - Solution

```python
from collections import defaultdict
class Solution:
    def isSelfCrossing(self, x):
        x_old = y_old = 0
        Direct = defaultdict(lambda : 0)
                
        Direct[(x_old, y_old)] += 1
        for i_, v_ in enumerate(x):
            for j_ in range(v_):
                if i_%4 == 0: y_old += 1
                elif i_%4 == 1: x_old -= 1
                elif i_%4 == 2: y_old -= 1
                elif i_%4 == 3: x_old += 1
                Direct[(x_old, y_old)] += 1
                if Direct[(x_old, y_old)] > 1:
                    return True
        return False
```

## Solution2

```python
class Solution:
    def isSelfCrossing(self, x) :
        for i in range(3,len(x)):
            if x[i-3]>=x[i-1] and x[i]>=x[i-2]:
                return True
            if i>=4:
                if x[i-3]==x[i-1] and x[i-2]<=x[i-4]+x[i]:
                    return True
            if i>=5:
                if x[i-4]<=x[i-2]<=x[i-4]+x[i] and x[i-1]<=x[i-3]<=x[i-5]+x[i-1]:
                    return True

        return False
```
