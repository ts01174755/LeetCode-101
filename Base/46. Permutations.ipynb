# 46. Permutations

Given an array nums of distinct integers, return all the possible permutations. You can return the answer in any order

## Solution

```python
class Solution:
    def permute(self, nums: List[int]) -> List[List[int]]:
        def func_(combs, nums, nums_len, res=[]):
            if len(nums) == 0: return
            for i in range(len(nums)):
                res.append(nums[i])
                func_(combs, nums[:i] + nums[i+1:], nums_len, res)
                if len(res) == nums_len:
                    combs.append(res[:])
                res.pop()
            return combs
        
        return func_([], nums, len(nums))
```
