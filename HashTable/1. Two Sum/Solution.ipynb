#### Solution 2 : dict (iter equal to n, but less memory usage)

```python
num_list = [3,3]
target=6
```

```python
def TwoSum_dic(nums, target):
    index = {}
    for i, x in enumerate(nums):
        if target - x in index:
            return [index[target - x], i]
        index[x] = i    
```

```python
TwoSum_dic(num_list,target)
```
